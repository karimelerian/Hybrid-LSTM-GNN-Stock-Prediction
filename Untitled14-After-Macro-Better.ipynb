Prepare LSTM Data

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim

# Concatenate stock data and macroeconomic data
combined_data = pd.concat([scaled_data, scaled_macro_data], axis=1)

# Create sequences for the combined data
def create_sequences(data, seq_length):
    xs, ys = [], []
    for i in range(len(data) - seq_length):
        x = data.iloc[i:(i + seq_length)].values
        y = data.iloc[i + seq_length].values[:len(stocks)]  # Target is still stock prices
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

seq_length = 21
X_combined, y = create_sequences(combined_data, seq_length)

# Convert to PyTorch tensors
X_combined_train = torch.tensor(X_combined, dtype=torch.float32)
y_train = torch.tensor(y, dtype=torch.float32)
```

Define LSTM Model

```python
class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size, dropout_prob=0.5):
        super(LSTM, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout_prob)
        self.fc = nn.Linear(hidden_size, output_size)
        self.dropout = nn.Dropout(dropout_prob)  # Additional dropout layer after LSTM

    def forward(self, x):
        batch_size = x.size(0)
        h0 = torch.zeros(self.num_layers, batch_size, self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, batch_size, self.hidden_size).to(x.device)

        out, _ = self.lstm(x, (h0, c0))
        out = self.dropout(out[:, -1, :])  # Apply dropout to the output of the LSTM
        out = self.fc(out)
        return out

# Define LSTM model
# Update input size for LSTM
input_size = X_combined_train.shape[2]  # Number of features (stock prices + macroeconomic indicators)
hidden_size = 50
num_layers = 2
output_size = y_train.shape[1]

lstm_model = LSTM(input_size, hidden_size, num_layers, output_size)
```

Prepare Graph Data

```python
import torch
!pip install torch-geometric
import torch_geometric
from torch_geometric.nn import GCNConv

# Initialize edges and edge weights
edges = []
edge_weights = []

# Add edges based on Pearson correlation (threshold > 0.7)
for i in range(len(stocks)):
    for j in range(i + 1, len(stocks)):
        corr_value = corr_matrix.iloc[i, j]
        if abs(corr_value) > 0.7:
            edges.append([i, j])
            edge_weights.append(corr_value)

# Add edges based on association rules (lift > 1.7)
from mlxtend.frequent_patterns import apriori, association_rules

stock_movements = (data.pct_change().dropna() > 0).astype(int)
frequent_itemsets = apriori(stock_movements, min_support=0.1, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.7)

for _, rule in rules.iterrows():
    i = stocks.index(next(iter(rule['antecedents'])))  # Extract first element
    j = stocks.index(next(iter(rule['consequents'])))  # Extract first element
    edges.append([i, j])
    edge_weights.append(rule['lift'])


# Convert to tensors
edge_index = torch.tensor(edges, dtype=torch.long).t().contiguous()
edge_weights = torch.tensor(edge_weights, dtype=torch.float32)

```

Define GNN Model

```python
class GNN(nn.Module):
    def __init__(self, num_features, hidden_size, num_classes, dropout_prob=0.5):
        super(GNN, self).__init__()
        self.conv1 = GCNConv(num_features, hidden_size)
        self.conv2 = GCNConv(hidden_size, num_classes)
        self.dropout = nn.Dropout(dropout_prob)  # Dropout layer

    def forward(self, x, edge_index, edge_weight=None):
        x = self.conv1(x, edge_index, edge_weight)
        x = torch.relu(x)
        x = self.dropout(x)  # Apply dropout after the first GCN layer
        x = self.conv2(x, edge_index, edge_weight)
        return x


# 🔹 Compute stock returns (used as node features for GNN)
returns = data.pct_change().dropna().values  # Shape: (num_samples, num_stocks)

# 🔹 Define GNN model
num_features = returns.shape[1]  # Each stock is a node, features = num_stocks
hidden_size = 16
gnn_output_size = num_features  # Output should match the number of stocks

gnn_model = GNN(num_features, hidden_size, gnn_output_size)

# 🔹 Convert stock returns to PyTorch tensor (used as GNN input)
X_gnn_train = torch.tensor(returns, dtype=torch.float32)  # Shape: (num_samples, num_stocks)

# 🔹 Define target labels (future stock returns for supervised learning)
y_train_gnn = torch.tensor(data.pct_change().shift(-1).dropna().values, dtype=torch.float32)

# 🔹 Ensure `y_train_gnn` is correctly aligned with `X_gnn_train`
y_train_gnn = y_train_gnn[:X_gnn_train.shape[0], :]  # Trim to match `X_gnn_train`
```

 Train GNN

```python
# Define optimizer
optimizer_gnn = optim.Adam(gnn_model.parameters(), lr=0.001)

# Training loop
for epoch in range(50):
    gnn_model.train()
    optimizer_gnn.zero_grad()

    # 🔹 Reshape X_gnn_train to (num_nodes, num_features) if needed
    X_gnn_train_flat = X_gnn_train.view(-1, X_gnn_train.shape[-1])

    # 🔹 Forward pass with edge weights
    gnn_outputs = gnn_model(X_gnn_train_flat, edge_index, edge_weights)

    # 🔹 Ensure outputs match y_train_gnn in shape
    gnn_outputs = gnn_outputs.view(y_train_gnn.shape)  # Reshape if needed

    # 🔹 Compute loss
    loss_gnn = torch.nn.functional.mse_loss(gnn_outputs, y_train_gnn)

    # 🔹 Backpropagation
    loss_gnn.backward()
    optimizer_gnn.step()

    # Print loss every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f'GNN Epoch [{epoch+1}/50], Loss: {loss_gnn.item():.6f}')
```

```python
class HybridModel(nn.Module):
    def __init__(self, lstm_model, gnn_model, lstm_output_size, gnn_output_size, final_output_size):
        super(HybridModel, self).__init__()
        self.lstm_model = lstm_model
        self.gnn_model = gnn_model
        self.fc = nn.Linear(lstm_output_size + gnn_output_size, final_output_size)

    def forward(self, x_lstm, x_gnn, edge_index):
        lstm_out = self.lstm_model(x_lstm)  # Shape: (batch_size, lstm_output_size)
        gnn_out = self.gnn_model(x_gnn, edge_index)  # Shape: (num_nodes, gnn_output_size)

        # Trim the GNN output to match the LSTM batch size
        gnn_out_trimmed = gnn_out[:x_lstm.size(0), :]  # Shape: (batch_size, gnn_output_size)

        # Concatenate LSTM and GNN outputs
        combined = torch.cat((lstm_out, gnn_out_trimmed), dim=1)  # Shape: (batch_size, lstm_output_size + gnn_output_size)

        # Pass through the fully connected layer
        out = self.fc(combined)  # Shape: (batch_size, final_output_size)
        return out

```

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define paths in Google Drive
drive_path = "/content/drive/My Drive/ColabModels"
os.makedirs(drive_path, exist_ok=True)  # Ensure the directory exists

model_path = os.path.join(drive_path, "hybrid_model.pth")
mse_file = os.path.join(drive_path, "mse_log.xlsx")

# Initialize model
hybrid_model = HybridModel(lstm_model, gnn_model, lstm_output_size, gnn_output_size, final_output_size)

"""def save_model(model, path=model_path):
    torch.save(model.state_dict(), path)
    print(f"Model saved to {path}")

def load_model(model, path=model_path):
    if os.path.exists(path):
        model.load_state_dict(torch.load(path))
        print("Model loaded successfully from Google Drive.")
    else:
        print("No saved model found, training from scratch.")

"""


def adjust_edge_index_for_batch(edge_index, batch_size):
    mask = (edge_index < batch_size).all(dim=0)
    return edge_index[:, mask]

print(f"X_lstm_train shape: {X_combined_train}")
print(f"X_gnn_train shape: {X_gnn_train.shape}")
print(f"y_train shape: {y_train.shape}")

def expanding_window_training(X_combined, X_gnn, y, model, edge_index, seq_length=21, epochs=50):
    """Train the model using an expanding window approach with model saving/loading."""
    #load_model(model, model_path)  # Load existing model if available

    train_size = int(len(X_combined) * 0.8)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.005)

    mse_data = []  # Store MSE values

    for i in range(train_size, len(X_combined) - seq_length):
        X_combined_train = torch.tensor(X_combined[:i], dtype=torch.float32)
        X_gnn_train = torch.tensor(X_gnn[:i], dtype=torch.float32)
        y_train = torch.tensor(y[:i], dtype=torch.float32)

        batch_size = X_combined_train.size(0)
        adjusted_edge_index = adjust_edge_index_for_batch(edge_index, batch_size)

        model.train()
        for epoch in range(epochs):
            optimizer.zero_grad()
            predictions = model(X_combined_train, X_gnn_train, adjusted_edge_index)
            loss = criterion(predictions, y_train)
            loss.backward()
            optimizer.step()

            if (epoch + 1) % 10 == 0:
                print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.6f}")

        # Test on next step
        model.eval()
        with torch.no_grad():
            X_combined_test = torch.tensor(X_combined[i:i+1], dtype=torch.float32)
            X_gnn_test = torch.tensor(X_gnn[i:i+1], dtype=torch.float32)
            y_test = torch.tensor(y[i:i+1], dtype=torch.float32)

            test_batch_size = X_combined_test.size(0)
            adjusted_edge_index_test = adjust_edge_index_for_batch(edge_index, test_batch_size)

            predictions_test = model(X_combined_test, X_gnn_test, adjusted_edge_index_test)
            mse = criterion(predictions_test, y_test).item()
            print(f"Day {i+1}, Test MSE: {mse:.6f}")

            mse_data.append([i+1, mse])

    #save_model(model, model_path)  # Save the trained model

    # Save MSE values to an Excel file
    mse_df = pd.DataFrame(mse_data, columns=["Day", "MSE"])
    mse_df.to_excel(mse_file, index=False)
    print(f"MSE log saved to {mse_file}")

# Call the function with combined data
expanding_window_training(X_combined_train, X_gnn_train, y_train, hybrid_model, edge_index, seq_length=21, epochs=50)


```
